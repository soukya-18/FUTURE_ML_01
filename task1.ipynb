```python
import pandas as pd

data = pd.read_csv("/Users/soukyabodire/Desktop/sales.csv")
print(data.head())
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data = data.sort_values('Date')
data['Day'] = range(len(data))
print(data.head())
```

```python
from sklearn.linear_model import LinearRegression

X = data[['Day']]
y = data['Sales']

model = LinearRegression()
model.fit(X, y)
```

```python
future_days = pd.DataFrame({'Day': [len(data)+i for i in range(10)]})
predictions = model.predict(future_days)

print(predictions)
```

```python
import matplotlib.pyplot as plt

plt.plot(data['Sales'], label='Actual')
plt.plot(range(len(data), len(data)+10), predictions, label='Forecast')
plt.legend()
plt.show()
```
