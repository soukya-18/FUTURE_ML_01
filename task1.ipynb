```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


data = pd.read_csv("/Users/soukyabodire/Desktop/sales.csv")

data['Date'] = pd.to_datetime(data['Date'])


data = data.sort_values('Date')

data['Day'] = range(len(data))

# Step 5: Prepare data
X = data[['Day']]
y = data['Sales']


model = LinearRegression()
model.fit(X, y)


future_days = pd.DataFrame({'Day': [len(data)+i for i in range(10)]})
predictions = model.predict(future_days)


print("Future Predictions:\n")
for i, val in enumerate(predictions):
    print(f"Day {len(data)+i} → Predicted Sales: {val:.2f}")


plt.figure(figsize=(10,5))

plt.plot(data['Sales'], label='Actual Sales', marker='o')
plt.plot(range(len(data), len(data)+10), predictions, label='Forecast Sales', linestyle='--')

plt.title("Sales Forecasting using Linear Regression")
plt.xlabel("Days")
plt.ylabel("Sales")

plt.legend()
plt.grid()

plt.show()
```
